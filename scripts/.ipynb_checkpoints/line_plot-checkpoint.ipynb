```python
def save_fig(filename, file_format='pdf', tight=True, **kwargs):
    if tight:
        plt.tight_layout()
    filename = filename.replace(' ', '_').lower()
    print('saved to ', filename)
    plt.savefig("/home/aszot/frl-master/data/figures/{}.{}".format(filename, file_format), format=file_format, dpi=1000, **kwargs)
    
full_colors = sns.hls_palette(n_colors=3, l=0.6, s=.9, h=0.7)
g_colors = {
    'push': full_colors[0],
    'obstacle': full_colors[1],
    'seesaw': full_colors[2],
    'train': full_colors[0],
    'test': full_colors[1],
}
    
def draw_line(log, method, num_data_points, ns, avg_step=3, mean_std=False, max_step=None, max_y=None, 
              ax=None, idx=0, debug_log=False):
    steps = {}
    values = {}
    seeds = log[method].keys()
#     print('seeds', seeds)
    for seed in seeds:
        step = np.array([x[0] for x in log[method][seed]])
        value = np.array([x[1] for x in log[method][seed]])
        
        if max_step:
            max_step = min(max_step, step[-1])
        else:
            max_step = step[-1]

        steps[seed] = step
        values[seed] = value
        if debug_log:
            print('%s data points: %i' % (seed, len(step)))
            print('%s max step: %i' % (seed, np.max(step)))

    data = []
    for seed in seeds:
        for i in range(len(steps[seed])):
            if steps[seed][i] <= max_step:
                data.append((steps[seed][i], values[seed][i]))
                
#     print('Total data ', len(data))
    data.sort()
    x_data = []
    y_data = []
    for step, value in data:
        x_data.append(step)
        y_data.append(value)
    x_data = np.array(x_data)
    y_data = np.array(y_data)
    
    n = len(x_data)
    # Smoothing factor
    x_data = x_data[:n // ns * ns].reshape(-1, ns)
    y_data = y_data[:n // ns * ns].reshape(-1, ns)
    x_data, y_data = np.mean(x_data, axis=1), np.mean(y_data, axis=1)

    n = len(x_data)
    # Number of data points
    avg_step = int(n // num_data_points)
    
    x_data = x_data[:n // avg_step * avg_step].reshape(-1, avg_step)
    y_data = y_data[:n // avg_step * avg_step].reshape(-1, avg_step)
    
    avg_x, avg_y = np.mean(x_data, axis=1), np.mean(y_data, axis=1)
    std_y = np.std(y_data, axis=1)

    min_y = np.min(y_data)
    max_y = np.max(y_data)
    
    
    
    l = ax.plot(avg_x, avg_y, label=g_common_name_map[method])
    plt.setp(l, linewidth=2, color=g_colors[method])
    ax.fill_between(avg_x, np.clip(avg_y - std_y, 0, max_y), np.clip(avg_y + std_y, 0, max_y), 
                    alpha=0.2, color=g_colors[method])#, facecolor=facecolor)


    return l, min_y, max_y
        
def draw_graph(logs, task, num_data_points, ns, xlabel='Step', ylabel='Success', legend=False, 
               put_title=True, mean_std=False, max_step=None, min_y=None, max_y=None, save_name='',
              title_font_size=22, ytic=None, ytic_factor=100.0):
    max_value = -999
    min_value = 999
    
    fig, ax = plt.subplots(figsize=(5, 4))
    
    lines = []
    num_colors = len(logs)
    for idx, method in enumerate(logs):
        seeds = logs[method].keys()
#         print('method: ', method, 'seed: ', seeds)
        if len(seeds) == 0:
            continue
        
        l_, min_, max_ = draw_line(logs, method, num_data_points, ns, 
                                   mean_std=mean_std, max_step=max_step, max_y=max_y, 
                                   ax=ax,
                                   idx=num_colors - idx - 1)
        lines += l_
        max_value = max(max_value, max_)
        min_value = min(min_value, min_)

    if min_y == None:
        min_y = int(min_value - 1)
    if max_y == None:
        max_y = max_value
        #max_y = int(max_value + 1)
#     print('min_y', min_y, 'max_y', max_y)
    if max_y == 1:
        plt.yticks(np.arange(min_y, max_y + 0.1, 0.2))
    else:
        if ytic is None:
            if max_y > 0.8:
                ytic = np.arange(0, 1.0, 0.2)
            elif max_y > 0.5:
                ytic = np.arange(0, 0.8, 0.2)
            elif max_y > 0.3:
                ytic = np.arange(0, 0.5, 0.1)
            elif max_y > 0.2:
                ytic = np.arange(0, 0.4, 0.1)         
            else:
                ytic = np.arange(0, 0.2, 0.05)
        if ytic_factor == 100.0:
            cv = lambda x: int(x)
        else:
            cv = lambda x: round(x, 1)
        print('using tics', ytic)
        plt.yticks(ytic, [cv(yt * ytic_factor) for yt in ytic])
            
    x_tic = np.arange(0, max_step + 0.1, int(max_step / 5))
    x_tic_name = ['%iM' % int((x / 1000000)) for x in x_tic]
    plt.xticks(x_tic, x_tic_name)
    ax.grid(b=True, which='major', color='lightgray', linestyle='--')

    ax.set_xlabel('Environment Steps', fontsize='large')
    plt.ylabel(ylabel, fontsize='large')
    ax.set_ylabel(ylabel, fontsize='large')
    
    ax.set_ylim(bottom=0)
    ax.set_xlim(0, max_step)
    
    if legend:
        labs = [l.get_label() for l in lines]
        plt.legend(lines, labs, fontsize='small', loc=0)
        #plt.legend()#bbox_to_anchor=(1.03, 0.73))

    if put_title:
        plt.title(task, fontsize=title_font_size)
    save_fig('learn/' + task.replace('/', '_') + '_' + save_name)

    
def get_wb_plot(run_name, scaling, use_field):
    api = wandb.Api()
    runs = api.runs('clvr/functional-rl', {"config.prefix": run_name})
    assert len(runs) == 1
    for wbrun in runs:         
        hist = wbrun.history(samples=15000)
        points = [(i * scaling, x) for i, x in enumerate(hist[use_field].tolist()) if x == x]
    return points
```

```python
g_common_name_map = {
    'main': 'Ours',
    'main2': 'Ours2',
    'tvae': 'VAE',
    'dist': 'Dist',
    'rnd_sample': 'RS',
    'no_ent': 'NE',
    'nn': 'NN',
    'fixed': 'FX',
    'ac_fixed': 'FX',
    'gt': 'GT',
    'im': 'IM',
    'ft': 'FT',
    
    'train': 'Train',
    'test': 'Test',
    
    'push': 'Push',
    'seesaw': 'Seesaw',
    'obstacle': 'Obstacle',
}

scaling = 256 * 32
task_ylabel = 'Success (%)'
task_key = 'eval_test_ep_goal_hit'

def load_task(task_name, keep_methods=['main', 'rnd_sample', 'no_ent', 'ac_fixed']):
    with open('/home/aszot/frl-master/analysis/eval_cmds/%s.yaml' % task_name) as f:
        methods = yaml.load(f)

    task = methods['task_name']

    del_keys = []
    for k in methods:
        if k.startswith('task_'):
            del_keys.append(k)

    for dk in del_keys:
        del methods[dk]

    methods = {k: v for k, v in methods.items() if k in keep_methods}
#     print(methods)

    
    return methods

def convert_to_logs(methods, task, use_field='eval_test_ep_goal_hit'):
    print('*** %s ***' % task)
    #use_methods = {k: v for k, v in methods.items() if k in ['main']}
    use_methods = methods
    logs = {}
    for method, seeds in use_methods.items():
        method_log = {}
        for seed in seeds:
            tmp_log = get_wb_plot(seed, scaling, use_field)
            print('Loaded ', seed)

    #         for k, v in tmp_log.items():
            method_log[seed] = tmp_log
        logs[method] = method_log
    return logs    
```

```python
ft_push_logs = convert_to_logs(load_task('push_ft', ['ft']), 'CREATE Push', 'ep_goal_hit')
ft_seesaw_logs = convert_to_logs(load_task('seesaw_ft', ['ft']), 'CREATE Seesaw', 'ep_goal_hit')
ft_obstacle_logs = convert_to_logs(load_task('obstacle_ft', ['ft']), 'CREATE Obstacle', 'ep_goal_hit')
```

```python
def extend_fill(logs, set_iter):
    max_count = 0
    for seed in logs:
        for it, val in logs[seed]:
            max_count = max(max_count, it)
    max_count = set_iter
    
    for seed in logs:
        if logs[seed][-1][0] < max_count:
            logs[seed].append((max_count, logs[seed][-1][1]))
    return logs

def max_out(log, until=10000000):
    max_val = 0
    max_i = -1
    for i, (it, x) in enumerate(log):
        if max_val < x:
            max_val = x
            max_i = i
    log = log[:max_i]
    
    cur_it = log[-1][0]
    while cur_it < until:
        log.append((cur_it, max_val))
        cur_it += 100000
    return log

ns = 2
num_data_points = 20
length = 5.5 * 1e6

plot = {k: max_out(v) for k,v in ft_push_logs['ft'].items()}
# max_out(plot['26-Push-71-IU-main'])

all_methods = {
    'push': {k: max_out(v) for k,v in ft_push_logs['ft'].items()},
    'seesaw': {k: max_out(v) for k,v in ft_seesaw_logs['ft'].items()},
    'obstacle': {k: max_out(v) for k,v in ft_obstacle_logs['ft'].items()},
}
draw_graph(all_methods, task='CREATE Fine Tune', num_data_points=num_data_points, ns=ns, 
               ylabel=task_ylabel, put_title=True, 
               legend=True, max_step=length, min_y=0, max_y=None, save_name='ft')
```

```python
nav_task_name = 'CREATE Push'
tasks = {
    'CREATE Push': (['push'], 'eval_test_ep_goal_hit'),
    'CREATE Seesaw': (['seesaw'], 'eval_test_ep_goal_hit'),
    'CREATE Obstacle': (['obstacle'], 'eval_test_ep_goal_hit'),
    'Shape Stacking': (['bs', 'bs_fixed'], 'eval_test_ep_final_height'),
    'Grid World': (['gw'], 'eval_test_ep_success'),
    'Recommender': (['reco'], 'eval_test_ep_avg_ctr'),
}



keep_methods=['main', 'no_ent', 'nn', 'fixed', 'dist', 'vae', 'fx']
for n, (load_files, test_metric) in tasks.items():
    all_methods = {}
    for load_f in load_files:
        methods = load_task(load_f, keep_methods=keep_methods)
        all_methods = {**all_methods, **methods}
    train_metric = test_metric.replace('test', 'train')
    train_logs = convert_to_logs(all_methods, n, use_field=train_metric)
    test_logs = convert_to_logs(all_methods, n, use_field=test_metric)
    task_logs[n] = (train_logs, test_logs)
```

```python


name_map = {
    'nn': 'Nearest Neighbor',
    'vae': 'Non-hierarchical VAE',
    'dist': 'Continuous Output',
    'no_ent': 'Ours w/o Entropy',
    'fixed': 'Ours w/o subsampling',
    'fx': 'Ours w/o subsampling',
    'main': 'Ours',
}
ss_def = np.arange(0.0, 2.5, 0.6)
plot_details = {
    'CREATE Push': (100.0, None, 'Success (%)'),
    'CREATE Seesaw': (100.0, None, 'Success (%)'),
    'CREATE Obstacle': (100.0, None, 'Success (%)'),
    'Shape Stacking': (1.0, {
        'nn': np.arange(0.0, 10.0, 2.0),
        'vae': ss_def,
        'dist': ss_def,
        'no_ent':ss_def,
        'fx': np.arange(0.0, 10.0, 2.0),
        'main': ss_def,
        'max_steps': 6000000,
        'num_dp': 25
    }, 'Height'),
    'Grid World': (100.0, None, 'Success (%)'),
    'Recommender': (1.0, None, 'CTR'),
}

def max_out(log, until=10000000):
    max_val = -1000
    max_i = -1
    for i, (it, x) in enumerate(log):
        if max_val < x:
            max_val = x
            max_i = i
    log = log[:max_i+1]
    
    cur_it = log[-1][0]
    while cur_it < until:
        log.append((cur_it, max_val))
        cur_it += 100000
    return log

def all_max_out(d, until):
    return {k: max_out(d[k], until) for k in d}

ns = 2
length = 6 * 1e7
print(nav_logs.keys())

for env, (train_methods, test_methods) in task_logs.items():
# env = 'Grid World'
# train_methods, test_methods = task_logs[env]
    factor, all_ytics, y_label = plot_details[env]
    for k in train_methods:
        if all_ytics is not None:
            ytic = all_ytics[k]
            max_steps = all_ytics['max_steps']
            num_data_points = all_ytics['num_dp']
        else:
            ytic = None
            max_steps = 60000000
            num_data_points = 50

        draw_graph({
            'train': all_max_out(train_methods[k], max_steps), 
            'test': all_max_out(test_methods[k], max_steps)}, 
                       task=env + ' ' + name_map[k], 
                       num_data_points=num_data_points, ns=ns, 
                       ylabel=y_label, put_title=True, 
                       legend=False, max_step=max_steps, min_y=0, max_y=None, save_name='learn',
                        title_font_size=16, ytic_factor=factor, ytic=ytic)
        plt.clf()
```
